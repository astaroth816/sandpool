```python
!pip install yfinance -q
import plotly.graph_objs as go
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.  pyplot as plt
import plotly.express as px
```

2.Download the data from yahoo(yfinance)

```python
Symbol = yf.Ticker('2376.TW')
start_time = '2021-01-01'
end_time = '2021-11-12'
Data = Symbol.history(period='1d',start=start_time, end=end_time, rounding=True)
Data.head()

```

3.Plot the Date/Close relation 

*Create the time_series*

```python
Data_time = Data.index.format(formatter=lambda x: x.strftime('%Y-%m-%d'))
print(Data_time)
Data['Close']
```

Filter the Data without data


```python
Data_1 = pd.DataFrame()
Data_1['Close'] = Data['Close']
Data_1['Low'] = Data['Low']
Data_1['High'] = Data['High']
```

Plot the graph

```python
trace0 = go.Scatter(x=Data_1.index, y=Data_1.Close, name='Close price' )
trace1 = go.Scatter(x=Data_1.index, y=Data_1.Low, name='Low')
trace2 = go.Scatter(x=Data_1.index, y=Data_1.High, name='High')
data = [trace0,trace1,trace2]
layout = go.Layout(title='The 2376.TW')
figure = go.Figure(data=data, layout=layout)
figure.update_layout(
    autosize=False,
    width=800,
    height=800,)
figure.show()
```

Plot the candle graph

Calculating the MA5,MA10,MA20,MA60,MA120,MA240

```python
#MA5
Data_1['MA5'] = Data_1['Close'].rolling(window=5).mean()
#MA10
Data_1['MA10'] = Data_1['Close'].rolling(window=10).mean()
#MA20
Data_1['MA20'] = Data_1['Close'].rolling(window=20).mean()
#MA60
Data_1['MA60'] = Data_1['Close'].rolling(window=60).mean()
#MA120
Data_1['MA120'] = Data_1['Close'].rolling(window=120).mean()
#MA240
Data_1['MA240'] = Data_1['Close'].rolling(window=240).mean()
```

Plot the Tech_Indicator

```python
trace0 = go.Scatter(x=Data_1.index, y=Data_1.Close, name='Close' )
trace1 = go.Scatter(x=Data_1.index, y=Data_1.MA5, name='MA5' )
trace2 = go.Scatter(x=Data_1.index, y=Data_1.MA10, name='MA10')
trace3 = go.Scatter(x=Data_1.index, y=Data_1.MA20, name='MA20')
trace4 = go.Scatter(x=Data_1.index, y=Data_1.MA60, name='MA60' )
trace5 = go.Scatter(x=Data_1.index, y=Data_1.MA120, name='MA120')
trace6 = go.Scatter(x=Data_1.index, y=Data_1.MA240, name='MA240')
data = [trace0,trace1,trace2,trace3,trace4,trace5,trace6]
layout = go.Layout(title='The 2376.TW')
figure = go.Figure(data=data, layout=layout)  
figure.show()
```

```python
def buying_and_selling(data): 
   signal_buy = []
   signal_sell = []
   flag =-1
   for i in range(len(Data_1)):
     if Data_1['MA5'][i] > Data_1['MA10'][i]:
       if flag != 1:
         signal_buy.append(Data_1['Close'][i])
         signal_sell.append(np.nan)
         flag = 1
       else:
         signal_buy.append(np.nan)
         signal_sell.append(np.nan)
     elif Data_1['MA5'][i] < Data_1['MA10'][i] :
       if flag != 0:
         signal_buy.append(np.nan)
         signal_sell.append(Data_1['Close'][i])
         flag = 0
       else:
         signal_buy.append(np.nan)
         signal_sell.append(np.nan)
     else:
         signal_buy.append(np.nan)
         signal_sell.append(np.nan)
   return (signal_buy , signal_sell)
```

```python
buying_and_selling = buying_and_selling(Data_1)
Data_1['buying'] = buying_and_selling[0]
Data_1['sell'] = buying_and_selling[1]

Data_1
```

```python
trace0 = go.Scatter(x=Data_1.index, y=Data_1.Close, name='Close')
trace1 = go.Scatter(x=Data_1.index, y=Data_1.MA5, name='MA5' )
trace2 = go.Scatter(x=Data_1.index, y=Data_1.MA10, name='MA10')
trace3 = go.Scatter(x=Data_1.index, y=Data_1.buying, name='buy signal',mode='markers', marker=dict(size=8,symbol='triangle-up', 
                              line=dict(width=0.5),color='red'))
trace4 = go.Scatter(x=Data_1.index, y=Data_1.sell, name='sell signal', mode='markers', marker=dict(size=8,symbol='triangle-down', 
                              line=dict(width=0.5),color='green'))
data = [trace0,trace1,trace2,trace3,trace4]
layout = go.Layout(title='The 2376.TW')
figure = go.Figure(data=data, layout=layout)
figure.show()
```
